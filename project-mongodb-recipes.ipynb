```python
# hide fields with None values
for key in list(recipe.keys()):
    if recipe[key] is None:
        del recipe[key]
recipe
```

```python
# extract values from single recipe
def extract_values(recipe):
    values = {
        'id': int(recipe['id']),
        'name': recipe['title'],
        'category': recipe['dishTypes'],
        'prepTime': int(recipe['readyInMinutes']),
        'source': recipe['sourceUrl'],
        'diets': recipe['diets'],
        'vegan': recipe['vegan'],
        'vegetarian': recipe['vegetarian'],
        'glutenFree': recipe['glutenFree'],
        'dairyFree': recipe['dairyFree'],
        'veryHealthy': recipe['veryHealthy'],
        'cuisines': recipe['cuisines'],
        'ingredients': [],
        'equipment': [],
    }
    if 'image' in recipe:
        values['image'] = recipe['image']
    else:
        values['image'] = placeholder

    if 'extendedIngredients' in recipe:
        ingredients_list = recipe['extendedIngredients']
        ingredient_names = [ingredient.get('name', '') for ingredient in ingredients_list]
        values['ingredients'] = ingredient_names
    
    if 'analyzedInstructions' in recipe:
        instructions = recipe['analyzedInstructions']
        for instruction in instructions:
            equipment_list = []
            for step in instruction.get('steps', []):
                equipment_list.extend([item['name'] for item in step.get('equipment', [])])
            values['equipment'].extend(equipment_list)
            
    return values
```

```python
# our random API allows max downoad of 100 random recipes at a time
# API is filtered by cuisine

list_recipes = []
num = 100

for i in range(0,len(cuisine_list)):
    # download recipes 
    url = "https://api.spoonacular.com/recipes/random?apiKey="+api_key+"&tags="+cuisine_list[i]+"&number="+str(num)
    r = requests.get(url)
    data = r.json()

    #add recipes to the list
    if data['recipes'] is not None:
        for recipe in data['recipes']:
            values = extract_values(recipe)
            list_recipes.append(values)
```

#### Checking and Removing Duplicate Recipes


Considering that our API provides a random assortment of recipes, it is anticipated that our database may contain duplicate entries. This arises from recipes being associated with multiple cuisine labels. In these instances, we may come across one or two duplicate recipes.

As we move forward, our focus will be on identifying these duplicates based on recipe IDs and subsequently removing them to ensure data integrity. This process will streamline our dataset for more effective analysis.


```python
# finding duplicate recipes
for d in recipes.find():
    print(d)
print()

pipeline = [
 {"$group": {
         "_id": {"id": "$id"} ,
         "count": { "$sum" : 1 }}},
 {"$match": {
         "count": { "$gt": 1 }}}]
```

```python
# counting duplicate recipes

for d in recipes.aggregate(pipeline):
    print(d)
print()
```

```python
# deleting duplicate recipes

pipeline = [
 {"$group": {
         "_id": {"id": "$id"} ,
         "duplicates": { "$addToSet": "$_id" } ,
         "count": { "$sum" : 1 }}},
 {"$match": {
         "count": { "$gt": 1 }}}]

for d in recipes.aggregate(pipeline):
    print(d)
    to_be_deleted = d["duplicates"][1:]  # all exept the first item should be deleted
    recipes.delete_many({
        "_id": {"$in": to_be_deleted}
    })
print()
```

```python
# counting duplicate recipes after the deletion duplicates

for d in recipes.aggregate(pipeline):
    print(d)
print()
```

```python
# count number of documents after deleting duplicates

recipes.count_documents({})
```

### Ranking Source Websites

In this chapter, we delve into the process of extracting and ranking source websites from a collection of diverse URLs in a MongoDB database. Using advanced aggregation techniques, we demonstrate how to effectively extract website names, filtering out prefixes such as 'http://' and 'https://.' We then proceed to rank these websites based on their frequency of occurrence, providing valuable insights into the most prominent sources within the dataset.

```python
# extracting the name of the website from the source url with regular expression
r = recipes.aggregate([
    {"$project": {"website": {"$regexFindAll": {"input": "$source","regex": "(?<=//)(.*?\\.com)"}}}},
    {"$unwind":"$website"},
    {"$group": {"_id": "$website.match","count": { "$sum": 1 }}},
    {"$project": {"_id": 0,"website": "$_id","count": 1}},
    {"$sort": {"count": -1}}])
pd.DataFrame(r)
```

```python
# plotting the resulty in a barchart
r = recipes.aggregate([
    {"$project": {"website": {"$regexFindAll": {"input": "$source","regex": "(?<=//)(.*?\\.com)"}}}},
    {"$unwind":"$website"},
    {"$group": {"_id": "$website.match","count": { "$sum": 1 }}},
    {"$project": {"_id": 0,"website": "$_id","count": 1}},
    {"$sort": {"count": -1}}])
# Create a bar chart
df=pd.DataFrame(r)
# Set the figure size to make the whole chart wider
fig, ax = plt.subplots(figsize=(10, 6))
# Create a bar chart using the 'websites' and 'count' columns
ax.bar(df['website'], df['count'], width=0.8)
# Add count numbers as text labels above each bar
for i, count in enumerate(df['count']):
    ax.text(i, count + 1, str(count), ha='center')
plt.xlabel('Websites')
plt.ylabel('Count')
plt.title('Source websites  of recipes')
plt.xticks(rotation=90)
# Show the plot
plt.tight_layout()
plt.show()
```

### Analysis of Dietary Categories by Cuisine

In the current phase, we are compiling a comprehensive list of all available diet types within the database. Subsequently, we are constructing a structured data frame to aggregate the number of recipes based on multiple factors, including cuisine, vegan, vegetarian, dairy-free, and gluten-free diets. This aggregation forms the foundation for a visually insightful bar and line chart, where the total count of recipes per cuisine is represented as bars, while the number of dietary categories is depicted as distinct colored lines. This graphical representation not only allows for easy comparison of dietary preferences across cuisines but also provides a dynamic view of the relationship between culinary traditions and dietary choices.

```python
# listing all diet types available in the database

r = recipes.aggregate([
    {"$project": { "diets":1}},
    {"$unwind": "$diets"},
    {"$group": {"_id": "$diets", "count": {"$sum": 1}}},
    {"$project": {"_id": 0, "diets":1,"name":"$_id", "count":1}},
    {"$sort": {"count": -1}},
 ])

pd.DataFrame(r)
```

```python
# Create data frame with aggregated recipe numbers

r = recipes.aggregate([
    # Unwind the cuisines array
    {"$unwind": "$cuisines"},
        # Group by cuisine and calculate the total count
    {"$group": {
        "_id": "$cuisines",
        "total_ct": {"$sum": 1},
        "vegan_ct": {"$sum": {"$cond": ["$vegan", 1, 0] }},
        "vegetarian_ct": {"$sum": {"$cond": ["$vegetarian", 1, 0] }},
        "dairyFree_ct": {"$sum": {"$cond": ["$dairyFree", 1, 0] }},
        "glutenFree_ct": {"$sum": {"$cond": ["$glutenFree", 1, 0] }}}},
        # Project the result to rename fields and exclude _id
    {"$project": {
        "_id": 0,
        "cuisine": "$_id",
        "total_ct": 1,
        "vegan_ct": 1,
        "vegetarian_ct": 1,
        "dairyFree_ct": 1,
        "glutenFree_ct": 1}},
        # Sort the results by total count in descending order
    {"$sort": {"total_ct": -1}}])
pd.DataFrame(r)
```

```python
# Plot aggregated table in a bar and line chart showing total count for the cuisines and number of diets as a colored line

r = recipes.aggregate([
    # Unwind the cuisines array
    {"$unwind": "$cuisines"},
    
    # Group by cuisine and calculate the total count
    {"$group": {
        "_id": "$cuisines",
        "total_ct": {"$sum": 1},
        "vegan_ct": {"$sum": {"$cond": ["$vegan", 1, 0] }},
        "vegetarian_ct": {"$sum": {"$cond": ["$vegetarian", 1, 0] }},
        "dairyFree_ct": {"$sum": {"$cond": ["$dairyFree", 1, 0] }},
        "glutenFree_ct": {"$sum": {"$cond": ["$glutenFree", 1, 0] }}}},
    
    # Project the result to rename fields and exclude _id
    {"$project": {
        "_id": 0,
        "cuisine": "$_id",
        "total_ct": 1,
        "vegan_ct": 1,
        "vegetarian_ct": 1,
        "dairyFree_ct": 1,
        "glutenFree_ct": 1}},
    
    # Sort the results by total count in descending order
    {"$sort": {"total_ct": -1}}])

df = pd.DataFrame(r)

# Set the figure size to adjust the plot dimensions
fig, ax = plt.subplots(figsize=(12, 6))

# Create a bar plot for "total_ct"
df.plot(kind="bar", x="cuisine", y="total_ct", ax=ax, label="total_ct")

# Plot lines for other columns
df.plot(kind="line", x="cuisine", y="vegan_ct", ax=ax, label="vegan_ct", secondary_y=True)
df.plot(kind="line", x="cuisine", y="vegetarian_ct", ax=ax, label="vegetarian_ct", secondary_y=True)
df.plot(kind="line", x="cuisine", y="glutenFree_ct", ax=ax, label="glutenFree_ct", secondary_y=True)
df.plot(kind="line", x="cuisine", y="dairyFree_ct", ax=ax, label="dairyFree_ct", secondary_y=True)

# Set labels and title
ax.set_xlabel("Cuisine")
ax.set_ylabel("Total Count (Bars)")
ax.set_title("Number of Dietary Categories by Cuisine")

# Show the legend
# Manually create a custom legend
custom_legend = [
    plt.Line2D([0], [0], color='b', label="vegan_ct"),
    plt.Line2D([0], [0], color='g', label="vegetarian_ct"),
    plt.Line2D([0], [0], color='r', label="glutenFree_ct"),
    plt.Line2D([0], [0], color='y', label="dairyFree_ct")
]
ax.legend(handles=custom_legend, loc="upper right")

# Show the plot
ax.set_xticklabels(df["cuisine"], rotation=90)
plt.tight_layout()
plt.show()
```

Notably, "main_dishes" emerges as the leader with more than 1000 entries, followed by "starters" at around 500. Other categories like "hor d'oeuvre," "snack," "side dish," and "dessert" are also well-represented. This consolidation of similar categories helps uncover valuable insights into the diverse array of dish categories available in the database.

```python
# listing all dish categories available in the database

r = recipes.aggregate([
    {"$project": { "category":1}},
    {"$unwind": "$category"},
    {"$group": {"_id": "$category", "count": {"$sum": 1}}},
    {"$project": {"_id": 0, "category":1,"name":"$_id", "count":1}},
    {"$sort": {"count": -1}},
    {"$limit": 10}
 ])

pd.DataFrame(r)
```

```python
# aggregating similar dish categories available in the database

r = recipes.aggregate([
    {"$project": { "category": 1 }},
    {"$unwind": "$category"},
    {"$group": {
            "_id": {
                "$switch": {
                    "branches": [
                        {"case": { "$in": ["$category", ["dinner", "lunch", "main dish", "main course"]] }, "then": "main_dishes" },
                        {"case": { "$in": ["$category", ["antipasti", "starter", "antipasto", "appetizer", "salad", "soup"]] }, 
                         "then": "starters" },
                        {"case": { "$in": ["$category", ["breakfast", "brunch", "morning meal"]] }, "then": "breakfast" },
                        {"case": { "$in": ["$category", ["drink", "beverage"]] }, "then": "drinks" },
                        {"case": { "$in": ["$category", ["sauce", "seasoning", "marinade", "dip"]] }, "then": "sauces" }],
                    "default": "$category"}},"count": {"$sum": 1}}},
    {"$project": {"_id": 0, "category": 1, "name": "$_id", "count": 1}},
    {"$sort": {"count": -1}}
])

pd.DataFrame(r)
```

### Analysis of Preparation Time

In our exploration of preparation time data, we first converted minutes to hours to facilitate a more intuitive analysis. We then delved into the 10 recipes with the longest and shortest preparation times, providing a glimpse of extreme cases. Subsequently, we visualized the distribution of recipes by preparation time, removing outliers to focus on the central trends. The data reveals a wide range of preparation times, with the majority falling within the typical cooking timeframe.

```python
# Showing the 10 recipes with the longest preparation time

r = recipes.aggregate([
    {"$project": {"name": 1, "id": 1, "display_image": 1, "prepTime": 1}},
    {"$addFields": {"prepTimeMinutes": {"$toInt": "$prepTime"}}},
    {"$addFields": {"PrepTimeHours": {"$divide": ["$prepTimeMinutes", 60] }}},
    {"$project": {"_id": 0, "id": 1, "name": 1, "prepTimeMinutes": 1, "PrepTimeHours": 1, "display_image": 1}},
    {"$sort": {"prepTimeMinutes": -1}},
    {"$limit": 5}])
df = pd.DataFrame(r)

# Display the DataFrame with HTML links and images
display(HTML(df.to_html(escape=False)))
```

```python
# Showing the 10 recipes with the shortest preparation time

r = recipes.aggregate([
    {"$project": {"name": 1, "id": 1, "display_image": 1, "prepTime": 1}},
    {"$addFields": {"prepTimeMinutes": {"$toInt": "$prepTime"}}},
    {"$addFields": {"PrepTimeHours": {"$divide": ["$prepTimeMinutes", 60] }}},
    {"$project": {"_id": 0, "id": 1, "name": 1, "prepTimeMinutes": 1, "PrepTimeHours": 1, "display_image": 1}},
    {"$sort": {"prepTimeMinutes": 1}},
    {"$limit": 5}])
df = pd.DataFrame(r)

# Display the DataFrame with HTML links and images
display(HTML(df.to_html(escape=False)))
```

```python
# Ploting the number of recipes by preparation time (including outliers)

r = recipes.aggregate([
    {"$project": { "prepTime": {"$divide": ["$prepTime", 60]} }},
    {"$group": {"_id": "$prepTime", "count": {"$sum": 1}}},
    {"$project": {"_id": 0, "prepTime": 1, "prepTime": "$_id", "count": 1}},
    {"$sort": {"prepTime": -1}}])

pd.DataFrame(r)
```

```python
# Ploting the number of recipes by preparation time after removing the outliers (All recipes which take longer than 1.5 h)

r = recipes.aggregate([
    {"$match": {"prepTime": {"$lt": 100}}},
    {"$project": { "prepTime": {"$divide": ["$prepTime", 60]} }},
    {"$group": {"_id": "$prepTime", "count": {"$sum": 1}}},
    {"$project": {"_id": 0, "prepTime": 1, "prepTime": "$_id", "count": 1}},
    {"$sort": {"prepTime": -1}}])

df=pd.DataFrame(r)
df.plot.line(x="prepTime",y="count")
```
